### 전처리 하기

- Unnamed: 0
    - 'index'로 칼럼 이름 변경
- ingredients
    - '\n Ingredients \n' 제거
    - ~~'\n' 대신 ','를 넣어 구분하기~~
    - '\n\n\nIngredients\n\n\n\nUnits\nUSM \n\nScale\n1x2x3x \n\n\n\n\n' 제거
- preparation
    - "* Top Tip *" 제거

**Unnamed: 0 전처리**

```python
df_recipe = df.rename(columns = {"Unnamed: 0": "index"})
```

```python
df_recipe = df_recipe.set_index('index')
df_recipe
```

### 전처리가 더 힘든 데이터 drop

```python
df_drop = df_recipe.drop([19, 116, 169, 177, 202, 487, 626, 757, 762])
df_drop
```

```python
df_drop_set = df_drop.reset_index(drop=True)
df_drop_set
```

```python
df_drop_set.shape
```

```python
df_drop_set2 = df_drop_set.drop([66,121,289,296,381,396,400,407,409,444,479,480,572,605,633,847])
```

```python
df_drop_set3 = df_drop_set2.reset_index(drop=True)
df_drop_set3
```

```python
df_drop_set3.shape
```

### ingredients 전처리

```python
import re
```

```python
j=[]
for i in range(1365):
    a = re.sub(r'\nIngredients\n', '', df_drop_set3["ingredients"][i]) # \nIngredients\n 삭제
    b = re.sub(r'\bScale\b', '', a) # \bScale\b 삭제
    c = re.sub(r'\b1x2x3x\b', '', b)# \b1x2x3x\b 삭제
    d = re.sub(r'\bUnits\b', '', c) # \bUnits\b 삭제
    e = re.sub(r'\bUSM\b', '', d)   # \bUSM\b 삭제
    f = re.sub(r'\n\n{1,}', '', e)  # 연속되는 \n 삭제
    g = re.sub(r'^\n', '', f)       # \n로 시작하는 것 삭제
    h = re.sub(' +(?= )', '', g)    # 연속되는 빈칸 삭제
    j.append(h)
    
df_drop_set3["new"] = j
df_drop_set3["new"]
```

```python
df_drop_set3
```

new 칼럼이 생겼으니 ingredients 칼럼 삭제

```python
df_drop_set4 = df_drop_set3.drop(["ingredients"], axis=1)
```

칼럼 이름 변경 (new -> ingredients)

```python
df_set4 = df_drop_set4.rename(columns = {"new": "ingredients"})
df_set4
```

### preparation 전처리
    - "Instructions" 제거   
    - "Method" or "Preparation" 제거   
    - '\xa0' 제거   
    - 연속된 공백, \n 제거

```python
j=[]
for i in range(1365):
    a = re.sub(r'\nMethod\n', '', df_set4["preparation"][i]) # \nMethod\n 삭제
    b = re.sub(r'\nPreparation\n', '', a)  # \nPreparation\n삭제
    c = re.sub(r'\nInstructions\n', '', b) # \nInstructions\n 삭제
    d = re.sub(r'\xa0', '', c)             # \xa0 삭제
    e = re.sub(r'\n\n{1,}', '', d)         # 연속되는 \n 삭제
    f = re.sub(r'^\n', '', e)              # \n로 시작하는 것 삭제
    j.append(f)

df_set4["new2"] = j
df_set4["new2"]
```

```python
df_set4
```

```python
df_set5 = df_set4.drop(["preparation"], axis=1)
```

```python
df_set6 = df_set5.rename(columns = {"new2": "preparation"})
```

href는 img_url이 아니지만 우선 MongoDB에 넣을 테스트를 하기 위해 컬럼 이름만 바꾸기

```python
df_set6 = df_set6.rename(columns = {"href": "img_url"})
df_set6
```

### 컬럼 정렬

```python
df_set7 = df_set6.reindex(["title", "ingredients", "preparation", "img_url"], axis=1)
df_set7
```

```python
df_vegan_recipe = df_set7.set_index('title')
df_vegan_recipe
```
