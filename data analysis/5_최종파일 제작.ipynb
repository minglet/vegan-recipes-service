### 완성된 similar csv 파일 업로드

```python
similar_df = pd.read_csv('./similar_recipe_837.csv')
similar_df
```

### df_recipe, similar_df merge하기

```python
final_df = pd.merge(df_recipe, similar_df, on='title', how='left')
final_df
```

### 837 -> 853개 중복된 값 검색

### ingredients를 기준으로 중복된 내용은 drop

```python
final_df = final_df.drop_duplicates(subset=['ingredients'])
final_df
```

### 정렬 및 정리

1. 인덱스 정렬
2. sim_recipe 제목에 맞는 인덱스 추출

```python
ffinal_df = final_df.reset_index(drop=True)
ffinal_df
```

### 레시피 제목이 아닌 인덱스 번호를 넣기

```python
sim_value = ffinal_df["sim_recipe"].tolist()

title_value = ffinal_df['title'].tolist()

# print(sim_value)
a = [] 
for i in sim_value:
    idx = title_value.index(i)
    a.append(idx)
ffinal_df["sim_idx"] = a
```
